```python
"""1.Expense Recording System

Write a Python program that allows the user to:
Enter daily expenses under categories: food, rent, travel, shopping, bills, others
Store them in a dictionary with category as key and amount as value
Print the total daily expense """

expenses = {}

categories = ["food", "rent", "travel", "shopping", "bills", "others"]

for cat in categories:
    amount = float(input(f"Enter today's expense for {cat}: "))
    
    expenses[cat] = amount

total = sum(expenses.values())

print("\nDaily Expenses:", expenses)
print("Total Daily Expense:", total)
```

```python
"""2. Monthly Expense Analyzer
Create a Python function that accepts a list of daily expenses for 30 days and returns:
Total monthly expense
Average daily expense
Maximum spending day
Minimum spending day"""

# Solution 2

def monthly_analysis(expense_list):
    total = sum(expense_list)
    average = total / len(expense_list)
    max_day = max(expense_list)
    min_day = min(expense_list)
    
    return total, average, max_day, min_day


# Example usage:
expenses = [2000, 550, 300, 700, 200, 650, 500] * 4  # Example 28-30 days
print(monthly_analysis(expenses))
```

```python
# """3. Conditional Savings Calculator
Write a program using if–else that:
Takes monthly income and monthly expense as input
Calculates savings = income − expense
Prints:
“Excellent Saver” if savings ≥ 40% of income
“Moderate Saver” if savings ≥ 20%
“Poor Saver” otherwise"""

income = float(input("Enter monthly income: "))
expense = float(input("Enter monthly expense: "))

savings = income - expense
percentage = (savings / income) * 100

if percentage >= 40:
    status = "Excellent Saver"
elif percentage >= 20:
    status = "Moderate Saver"
else:
    status = "Poor Saver"

print("\nSavings:", savings)
print("Category:", status)
```

```python
""" 5. Function-Based Budget Advice
Write a function:
The function should return a recommendation:
If rent > 40% of income → "Reduce rent"
If food > 30% → "Control food spending"
If transport > 20% → “Transport cost high”
Else → “Good budget management” """


def budget_advice(income, rent, food, transport):
    if rent > 0.40 * income:
        return "Reduce rent"
    elif food > 0.30 * income:
        return "Control food spending"
    elif transport > 0.20 * income:
        return "Transport cost high"
    else:
        return "Good budget management"


print(budget_advice(5000, 2000, 1000, 6000))
```

```python
""" 6. Track Recurring Expenses Using Tuples & Sets

Create two tuples:
Tuple A = recurring fixed expenses (rent, WiFi, subscriptions)
Tuple B = recurring variable expenses (food, travel, shopping)
Tasks:
Convert both to sets
Identify common items
Identify items unique to each category"""

fixed_expenses = ("rent", "wifi", "subscriptions")
variable_expenses = ("food", "travel", "shopping", "subscriptions")

set_fixed = set(fixed_expenses)
set_variable = set(variable_expenses)

common_items = set_fixed & set_variable
unique_fixed = set_fixed - set_variable
unique_variable = set_variable - set_fixed

print("Common items:", common_items)
print("Unique Fixed:", unique_fixed)
print("Unique Variable:", unique_variable)
```

```python
""" 7. Class-Based Financial Profile
Create a class FinanceProfile with attributes:
name
monthly_income
total_expense
Include methods:
calculate_savings()
spending_status() returning: “Saving”, “Balanced”, or “Overspending”
Create objects for 3 users and print their status."""


class FinanceProfile:
    def __init__(self, name, income, expense):
        self.name = name
        self.income = income
        self.expense = expense

    def calculate_savings(self):
        return self.income - self.expense

    def spending_status(self):
        s = self.calculate_savings()
        if s > 0:
            return "Saving"
        elif s == 0:
            return "Balanced"
        else:
            return "Overspending"


p1 = FinanceProfile("Vijay", 5000, 4200)
p2 = FinanceProfile("Ramesh", 3500, 3500)
p3 = FinanceProfile("Ajith", 6000, 7500)

print(p1.name, "→", p1.spending_status())
print(p2.name, "→", p2.spending_status())
print(p3.name, "→", p3.spending_status())
```

```python
""" 8. File Handling: Save & Load Expense Logs
Write a program to:
Accept user expenses for 7 days
Save them to a file expenses.txt
Then load the file and print the total weekly expense
Add error handling for missing or unreadable files."""


weekly_expenses = [500, 300, 450, 700, 200, 550, 600]

try:
    with open("expenses.txt", "w") as f:
        for e in weekly_expenses:
            f.write(str(e) + "\n")
    print("File saved successfully.")

except:
    print("Error saving file.")


try:
    total = 0
    with open("expenses.txt", "r") as f:
        for line in f:
            total += float(line.strip())
    print("Total Weekly Expense:", total)

except FileNotFoundError:
    print("File not found.")

except:
    print("Error reading file.")

finally:
    print("Program Completed.")
```

```python
"""9.Create a Python class named ExpenseTracker that helps manage a user's personal finance details.

The class must:

Use an __init__() function to initialize the following attributes:
name,monthly_income,rent,food,transport,others
Include a method total_expense() that returns the sum of all expenses.
Include a method savings() that returns:
monthly_income – total_expense
Include a method status() that returns:
"Saving" if savings > 0
"Balanced" if savings == 0
"Overspending" if savings < 0
Create an object of this class and print:
Total Expense,Savings,Status"""

class ExpenseTracker:
    def __init__(self, name, monthly_income, rent, food, transport, others):
        self.name = name
        self.monthly_income = monthly_income
        self.rent = rent
        self.food = food
        self.transport = transport
        self.others = others

    def total_expense(self):
        return self.rent + self.food + self.transport + self.others

    def savings(self):
        return self.monthly_income - self.total_expense()

    def status(self):
        s = self.savings()
        if s > 0:
            return "Saving"
        elif s == 0:
            return "Balanced"
        else:
            return "Overspending"


# Creating an object
person1 = ExpenseTracker("Ava", 5000, 2000, 1200, 500, 300)

print("Total Expense:", person1.total_expense())
print("Savings:", person1.savings())
print("Status:", person1.status())
```

```python
'''10.Write a Python program :

Take monthly income and different expenses from the user
Calculate the total expense,Calculate savings
Display whether the user is Saving, Balanced, or Overspending
Use separate functions for:
Calculating total expense
Checking spending status'''

def calculate_total_expense(rent, food, transport, others):
    return rent + food + transport + others


def calculate_status(income, total_expense):
    savings = income - total_expense
    
    if savings > 0:
        return "Saving"
    elif savings == 0:
        return "Balanced"
    else:
        return "Overspending"


def main():
    print("=== Personal Finance Tracker ===")
    
    income = float(input("Enter your monthly income: "))
    rent = float(input("Enter rent expense: "))
    food = float(input("Enter food expense: "))
    transport = float(input("Enter transport expense: "))
    others = float(input("Enter other expenses: "))
    
    total_expense = calculate_total_expense(rent, food, transport, others)
    status = calculate_status(income, total_expense)
    
    print("\n--- Financial Summary ---")
    print("Total Expense:", total_expense)
    print("Savings:", income - total_expense)
    print("Status:", status)

if __name__ == "__main__":
    main()
```
